```python
import tkinter

#MVC Architecure

#Controller
def click(var,value):
    var.set(var.get() + value)

if __name__ == '__main__':
    window = tkinter.Tk()
    
    #Model
    counter = tkinter.IntVar()
    counter.set(0)
    
    #Views
    frame = tkinter.Frame(window)
    frame.pack()
    
    button = tkinter.Button(frame,text = 'Increment',command = lambda:click(counter,1))  
    button.pack()
    button = tkinter.Button(frame,text = 'Decrement',command = lambda:click(counter,-1))  
    button.pack()
    
    label = tkinter.Label(frame,textvariable = counter)
    label.pack()
       
window.mainloop()
```

```python
import tkinter

# '#00000 to #FFFFFF' for color codes
# first 2 for red,next 2 for green,last 2 for blue
# Red = '#FF0000'
# Green = '#00FF00'
# Blue = '#0000FF'

def recolor(widget,r,g,b):
    color = '#'
    for var in (r,g,b):
        color+='FF' if var.get() else '00'
    widget.config(bg = color)
window = tkinter.Tk()
frame = tkinter.Frame(window)
frame.pack()

red = tkinter.IntVar()
green = tkinter.IntVar()
blue = tkinter.IntVar()

for (name,var) in (('R',red),('G',green),('B',blue)):
    check = tkinter.Checkbutton(frame,text = name,variable = var)
    check.pack(side = 'left')
    
label = tkinter.Label(frame,text = '[     ]')
button = tkinter.Button(frame,text = 'Update',command = lambda:recolor(label,red,green,blue))
button.pack(side = 'left')
label.pack(side = 'left')

window.mainloop()
```

```python
import tkinter
import tkinter.filedialog as dialogue

def quit(root):
    root.destroy()
def save(root,text):
    data = text.get('0.0',tkinter.END)
    filename = dialogue.asksaveasfilename(
    parent = root,
    filetype = [('Text','*.txt')],
    title = 'Save As....')
    
    writer = open(filename,'w')
    writer.write(data)
    writer.close()
    
window = tkinter.Tk()

text = tkinter.Text(window)
text.pack()

menubar = tkinter.Menu(window)
filemenu = tkinter.Menu(menubar)

filemenu.add_command(label = 'Save',command = lambda:save(window,text))
filemenu.add_command(label = 'Quit',command = lambda:quit(window))

menubar.add_cascade(label = 'File',menu = filemenu)
window.config(menu = menubar)

window.mainloop()
```

```python
import tkinter as tk

class Counter:
    """A program for GUI using OOPs."""
    def __init__(self,parent):
        """Create a GUI."""
        
        #FRamework
        self.parent = parent
        self.frame = tk.Frame(parent)
        self.frame.pack()
        
        #
        self.state = tk.IntVar()
        self.state.set(1)
        
        #Label
        self.label = tk.Label(self.frame,textvariable = self.state)
        self.label.pack()
        
        #Button 1
        self.button = tk.Button(self.frame,text = 'exit',command = self.exit1)
        self.button.pack()
        
        #Button 2
        self.button1 = tk.Button(self.frame,text = 'increment',command = self.click)
        self.button1.pack()
        
    def click(self):
        """For clicking increment button"""
        self.state.set(self.state.get()+1)
        
    def exit1(self):
        """Exit the window"""
        self.parent.destroy()
if __name__ == '__main__':
    window = tk.Tk()
    myapp = Counter(window)
    window.mainloop()
```
